### Updating global settings...

 git config --global user.name "YourUserName"

 git config --global user.email yourEmailID_G_Hub

 git config --global init.default branch main

 ### cd into the required location...

 git init 

 #### To initialize a new repository at the required location...

 git status

 #### Displays the current status of the file...

 git add .  

 #### Adds all files in the working directory to the staging area...

 git add filename

 #### Adds only one specific file to the staging area...
 
 git rm --cached filename

 #### Removes a file from the staging area but keeps it in the working directory...

 ### .gitignore
 If you want to ignore certain files or directories...

 ' # ' acts as a comment...
 example:
 *.txt
 filename.extension
 ignores all text files in your local repo...

 ### Committing changes to the repository...
 start with :

 git status 

 to check which are all the staged changes...
 
 git add .

 git commit -m "Some message"
 
 ### Post commit...
 You can view history using...
 git log

 any more changes done... the diffrences can be viewed with 
 git diff 

 ### To remove any staged changes without deleting the file we use:
 git restore --staged filename

 ### option exists to bypass staging...
 git commit -a -m "message..."

 ### to remove a file...
 git rm "filename"

 ### to restore it...
 git restore "filename"
